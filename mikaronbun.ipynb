```python
import numpy as np
import pandas as pd

pd.options.mode.use_inf_as_na = True
```

```python
df = pd.read_excel("CSR_koyou_2022.北島入力xlsx.xlsx")
```

```python
colums_list = [
    "環境(5)",
    "企業統治(5)",
    "社会性(5)",
    "30歳平均賃金",
    "月平均残業時間",
    "有休取得率/前期",
    "従業員の能力・評価結果の公開",
    "勤続年数合計/2期前",
    "3年前4月1日入社・合計(入社者定着状況)",
    "うち直近4月1日在籍者・合計(入社者定着状況)",
]
df1 = df.loc[:, colums_list]
df1
```

```python
df2 = df1.dropna()
df2 = df2[df2["3年前4月1日入社・合計(入社者定着状況)"] != 0]
df2
```

```python
df2.to_csv("data.csv")
```

```python
df3 = (
    df2["3年前4月1日入社・合計(入社者定着状況)"]
    - df2["うち直近4月1日在籍者・合計(入社者定着状況)"]
) / df2["3年前4月1日入社・合計(入社者定着状況)"]
df3.name = "3年以内離職率"
df3
```

```python
df3.to_csv("3年以内離職率")
```

```python
df5 = df2.iloc[:, :-2]
df5
```

```python
df4 = pd.concat((df5, df3), axis=1)
df4.dropna()
df4
```

```python
df4.to_csv("data2.csv")
```
